# 12 - Number of companies founder founded before

```python
import pandas as pd
import numpy as np
```

```python
companies = pd.read_csv("../raw_data/companies.csv")
```

```python
people = pd.read_csv("../raw_data/people.csv")
```

```python
relationships = pd.read_csv("../raw_data/relationships.csv")
```

```python
founders = pd.read_csv("../raw_data/founders.csv")
```

```python
founders.head()
```

```python
founding = relationships[relationships.founder]
```

```python
founding = founding.rename(columns={"relationship_object_id":"id"})
```

```python
founding.shape
```

```python
founding.head()
```

```python
companies[["id","founded_at"]].head()
```

```python
merged = companies.merge(founding, how="left", on="id")
```

```python
merged.founded_at = pd.to_datetime(merged.founded_at)
```

```python
merged.shape
```

```python
merged.head()
```

```python
# Number of companies founded in total
```

```python
companies_founded = merged[["person_object_id","founder"]].groupby("person_object_id"\
                    ,as_index=False).sum().rename(columns={"founder":"founded_companies"})
```

```python
companies_founded.head()
```

```python
companies_founded.founded_companies = companies_founded.founded_companies.astype("int")
```

```python
founders = founders.merge(companies_founded, how="left", on="person_object_id")
```

```python
founders.head()
```

```python
companies_founded.head()
```

```python
tmp_2 = merged.merge(companies_founded, how="left", on="person_object_id")[["id","founded_companies"]].groupby("id", as_index=False).mean()
```

```python
tmp_2 = tmp_2.rename(columns={"founded_companies":"mean_comp_founded_ever"})
```

```python
companies = companies.merge(tmp_2)
```

```python
## order of companies
```

```python
merged.shape
```

```python
merged.head()
```

```python
tmp = merged.sort_values(by=["person_object_id","founded_at"]).groupby("person_object_id").cumcount()
```

```python
tmp = pd.concat([merged, tmp], axis=1).sort_values(by=["person_object_id","founded_at"])
```

```python
tmp = tmp.rename(columns={0:"founded_count"})
```

```python
tmp.loc[tmp.person_object_id.isnull(),'founded_count']=np.nan
```

```python
tmp = tmp[['id',"founded_count"]].groupby("id",as_index=False).mean().rename(columns={"founded_count":"mean_comp_founded_before"})
```

```python
tmp.shape
```

```python
tmp.head()
```

```python
companies.shape
```

```python
companies.head()
```

```python
companies = companies.merge(tmp, how="left", on="id")
```
