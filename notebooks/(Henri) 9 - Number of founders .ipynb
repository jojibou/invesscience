# keeping founders 

```python
founders = df['founder']==True
```

```python
founders.head()
```

```python
df_filter = df[founders]
print(df_filter.shape)
```

```python
df_filter
```

# group by company id and count 

```python
founder_count = df_filter.groupby('relationship_object_id', as_index = False )['person_object_id'].count()


```

```python
founder_count
```

```python
founder_count.rename({"relationship_object_id": "company_id"}, inplace=True)
```

# Left merge to company table 

```python
companies = pd.read_csv("/Users/henrilehodey/code/jojibou/invesscience/raw_data/companies.csv")
companies.head()
```

```python
companies.rename({"id": "company_id"})
```

```python
how_many_founders = pd.merge(companies, founder_count, how='left', on='company_id')


```
